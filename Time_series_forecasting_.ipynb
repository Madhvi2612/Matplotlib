```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Monthly sales
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [100, 120, 130, 140, 150, 160]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the average increase in sales
df['Increase'] = df['Sales'].diff()
average_increase = df['Increase'][1:].mean()

# Forecast next three months
last_sales = df['Sales'].iloc[-1]
forecast = [last_sales + (i + 1) * average_increase for i in range(3)]
# Create forecast DataFrame
forecast_months = ['July', 'August', 'September']
forecast_df = pd.DataFrame({'Month': forecast_months, 'Forecasted Sales': forecast})

# Plot the data
plt.plot(df['Month'], df['Sales'], label='Actual Sales', marker='o')
plt.plot(forecast_months, forecast, label='Forecasted Sales', linestyle='--', marker='o', color='red')
plt.title("Sales Forecast")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()

# Display forecasted values
print("Forecasted Sales for the next 3 months:")
print(forecast_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Monthly sales
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [100, 120, 130, 140, 150, 160]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the average increase in sales
df['Increase'] = df['Sales'].diff()
average_increase = df['Increase'][1:].mean()

# Forecast next three months
last_sales = df['Sales'].iloc[-1]
forecast = [last_sales + (i + 1) * average_increase for i in range(6)]
# Create forecast DataFrame
forecast_months = ['July', 'August', 'September','October','November','December']
forecast_df = pd.DataFrame({'Month': forecast_months, 'Forecasted Sales': forecast})

# Plot the data
plt.plot(df['Month'], df['Sales'], label='Actual Sales', marker='o')
plt.plot(forecast_months, forecast, label='Forecasted Sales', linestyle='--', marker='o', color='red')
plt.title("Sales Forecast")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()

# Display forecasted values
print("Forecasted Sales for the next 3 months:")
print(forecast_df)
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Sample time series data
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'Sales': [200, 220, 250, 270, 300, 330]}
df = pd.DataFrame(data)
df.index = pd.date_range(start='2023-01', periods=len(df), freq='ME')

# Fit ARIMA model (p=1, d=1, q=0)
model = ARIMA(df['Sales'], order=(1, 1, 0))
fitted_model = model.fit()

# Forecast for the next 3 months
forecast = fitted_model.forecast(steps=5)

# Plot original data and forecast
plt.plot(df.index, df['Sales'], label='Actual Sales', marker='o')
future_dates = pd.date_range(start=df.index[-1], periods=6, freq='ME')[1:]
plt.plot(future_dates, forecast, label='Forecasted Sales', linestyle='--', marker='o', color='red')
plt.title("Sales Forecast with ARIMA")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.legend()
plt.show()

# Display forecasted values
print("Forecasted Sales:")
print(forecast)
```

SARIMA

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Sample time series data: Monthly sales
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
        'Sales': [200, 220, 250, 270, 300, 330, 350, 370, 400, 420, 450, 480]}
df = pd.DataFrame(data)
df.index = pd.date_range(start='2023-01', periods=len(df), freq='M')

# Fit SARIMA model (p, d, q)(P, D, Q, m)
model = SARIMAX(df['Sales'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
fitted_model = model.fit()

# Forecast for the next 6 months
forecast = fitted_model.forecast(steps=6)

# Plot original data and forecast
plt.plot(df.index, df['Sales'], label='Actual Sales', marker='o')
future_dates = pd.date_range(start=df.index[-1], periods=7, freq='M')[1:]
plt.plot(future_dates, forecast, label='Forecasted Sales', linestyle='--', marker='o', color='red')
plt.title("Sales Forecast with SARIMA")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()

# Display forecasted values
print("Forecasted Sales:")
print(forecast)
```

```python
from pmdarima import auto_arima
import pandas as pd

# Sample data: Monthly sales
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'Sales': [200, 220, 250, 270, 300, 330]}
df = pd.DataFrame(data)
df.index = pd.date_range(start='2023-01', periods=len(df), freq='ME')

# Auto ARIMA to find the best model
model = auto_arima(df['Sales'], seasonal=True, n=12, trace=True)

# Summary of the best model
print(model.summary())
```
